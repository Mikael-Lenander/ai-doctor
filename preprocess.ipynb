```python
import pandas as pd
import kagglehub
from pathlib import Path
```

```python
path = Path(kagglehub.dataset_download("artemminiailo/medicalconversations2disease")) / "medical_conversations.csv"
pd.set_option("display.max_colwidth", 1000)
df = pd.read_csv(path, names=('text', 'label'), skiprows=1)
df.head()
```

```python
def generate_template(conversation):
    roles = {
        "user": "user",
        "bot": "assistant"
    }
    lines = conversation.split('</s>')
    template = [{
        "role": "system",
        "content": "You are an empathetic doctor chatbot that asks follow-up questions about the patient's symptoms and provides medical guidance."
    }]
    for line in lines:
        sep_idx = line.find(":")
        if sep_idx == -1:
            continue
        message = {
            "role": roles[line[:sep_idx].lower().strip()],
            "content": line[sep_idx+1:].strip()
        }
        template.append(message)

    return template

chat_templates = list(map(generate_template, df['text'].tolist()))
chat_templates
```
